# WhatsApp Chat Analysis 

As usual We'll preprocess the data first

```python
import re 
import pandas as pd
```

```python
chats = open("whatsapp chats data.txt" ,'r' , encoding ="utf-8")
```

```python
data = chats.read()
```

```python
print(data)
```

```python
pattern = r'\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}'


```

```python
messages = re.split(pattern , data)[1:]
print(messages)
```

```python
dates = re.findall(pattern , data)
dates
```

```python
#  now we'll convert all of this into a dateframe 
```

```python
df = pd.DataFrame({'user_message':messages , "message_date":dates})
df['message_date'] = pd.to_datetime(df['message_date'] , format='%d/%m/%Y, %H:%M')
df.rename(columns = {'message_date':'date'} , inplace = True)
df.head()
```

```python
import re

# Define a function to clean the "am - " or "pm - " prefix
def clean_am_pm(text):
    return re.sub(r'^\s*(am|pm)\s*-\s*', '', text, flags=re.IGNORECASE)

# Apply the function to all rows in the 'user_message' column
df['user_message'] = df['user_message'].apply(clean_am_pm)


print(df.head())  

```

```python
df
```

```python
df.shape
```

```python
users =[]
messages =[]
for message in df['user_message']:
    entry = re.split('([\w\W]+?):\s', message)
    if entry[1:]:
        users.append(entry[1])
        messages.append(entry[2])
    else:
        users.append('group_notification')
        messages.append(entry[0])
```

```python
df['user'] = users
df['message'] = messages
df.drop(columns = ['user_message'] , inplace = True)
df
```

```python
df['year'] =df['date'].dt.year
df['month'] = df['date'].dt.month_name()
df['day'] = df['date'].dt.day
df['hour'] = df['date'].dt.hour     
df['minute'] = df['date'].dt.minute
df
```
