# WhatsApp Chat Analysis 

As usual We'll preprocess the data first

```python
import re 
import pandas as pd
```

```python
chats = open("whatsapp chats data.txt" ,'r' , encoding ="utf-8")
```

```python
data = chats.read()
```

```python
print(data)
```

```python
# pattern = '\d{1,2}/\d{1,2}/\d{2,4},\s\d{1,2}:\d{2}\s-\s'

pattern = '\d{2}/\d{2}/\d{4}, \d{1,2}:\d{2}\'
```

```python
messages = re.split(pattern , data)[1:]
print(messages)
```

```python
dates = re.findall(pattern , data)
dates
```
